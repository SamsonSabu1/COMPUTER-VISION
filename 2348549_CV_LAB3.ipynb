```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/onepiece1.jpg', 0)

#  Find width and height of the image
row, column = img.shape

#  Create a zeros array to store the sliced image
img1 = np.zeros((row, column), dtype='uint8')

#  Specify the min and max range
min_range = 80
max_range = 140

# Loop over the input image and if pixel value lies in the desired range set it to 255
# otherwise set it to the desired value
for i in range(row):
    for j in range(column):
        if min_range < img[i, j] < max_range:
            img1[i, j] = 255
        else:
            img1[i, j] = img[i - 1, j - 1]

# Save the images
cv2.imwrite('/content/Original.jpg', img)
cv2.imwrite('/content/slicedimage.jpg', img1)

# Display the original and sliced images
cv2_imshow(img)
cv2_imshow(img1)
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the input image
image = cv2.imread('/content/doggy.jpg', 0)  # Read the image in grayscale

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original image
cv2_imshow(image)

# Display the equalized image
cv2_imshow(equalized_image)

```
