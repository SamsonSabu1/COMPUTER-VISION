Read, write and view the images using different python libraries (CV2, ImageIO, MatplotLib, PIL)

```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread("/content/cowboybepop.jpg")
cv2_imshow(img)
```

```python
cv2.imwrite("g3g.png",img)
```

```python
import imageio as iio
img=iio.imread("/content/cowboybepop.jpg")
iio.imwrite("g4g.png",img)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
Image = img.imread('/content/cowboybepop.jpg')
plt.imshow(Image)
```

```python
from PIL import Image
```

```python
image = Image.open("/content/cowboybepop.jpg")
```

```python
image.save('new_image.jpg')
```
