```python
In this Python program, 
we load an image as grayscale, 
define two points corresponding to input and output images, 
get the transformation matrix, 
and finally apply the warpAffine() method to perform affine transformation on the input image.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
Read the input image using cv2.imread() function. Pass the full path of the input image.
```

```python
img = cv2.imread('C:/Users/samso/Music/lines.jpg', 0)
```

```python
retrieves the number of rows and columns in the input image 'img' using the shape attribute.
```

```python
rows,cols = img.shape
```

```python
Two sets of points (pts1 and pts2) are created as numpy arrays of type float32. 
These points will be used to define the transformation from pts1 to pts2.
```

```python
pts1 = np.float32([[150,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[10,250]])
```

```python
The getAffineTransform function calculates the 2x3 matrix M required for the affine transformation, 
based on the correspondence between pts1 and pts2.
```

```python
The warpAffine function applies the affine transformation to the input image img using the transformation matrix M to produce the output image dst.
```

```python
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
The imshow function is used to display the images, and subplot is used to arrange them in a grid for visualization. 
The show function then displays the plot to the use
```

```python
plt.subplot(121), plt.imshow(img, cmap='gray' ), plt.title('Input')
plt.subplot(122), plt.imshow(dst, cmap='gray'), plt.title('Output')
plt.show()
```

```python
What is Affine Transformation
```

```python
In Affine transformation, all parallel lines in the original image will still be parallel in the output image.
```

```python
To find the transformation matrix, we need three points from input image and their corresponding locations in the output image. 
```

```python
Then cv2.getAffineTransform will create a 2×3 matrix which is to be passed to cv2.warpAffine.
```

```python
cv2.getAffineTransform method:

Syntax: cv2.getPerspectiveTransform(src, dst)
Parameters: src: Coordinates of quadrangle vertices in the source image. 
dst: Coordinates of the corresponding quadrangle vertices in the destination image.
```

```python
cv2.warpAffine method:

Syntax: cv2.warpAffine(src, M, dsize, dst, flags, borderMode, borderValue) 
Parameters: src: input image. 
dst: output image that has the size dsize and the same type as src.
M: transformation matrix. 
dsize: size of the output image. 
flags: combination of interpolation methods (see resize() ) and the optional flag WARP_INVERSE_MAP that means that M is the inverse transformation (dst->src). 
borderMode: pixel extrapolation method; when borderMode=BORDER_TRANSPARENT, it means that the pixels in the destination image corresponding to the “outliers” in the source image are not modified by the function. 
borderValue: value used in case of a constant border; by default, it is 0.
```

```python
What is an Affine Transformation?
```

```python
A transformation that can be expressed in the form of a matrix multiplication (linear transformation) followed by a vector addition (translation).
From the above, we can use an Affine Transformation to express:

Rotations (linear transformation)
Translations (vector addition)
Scale operations (linear transformation)
you can see that, in essence, an Affine Transformation represents a relation between two images.

The usual way to represent an Affine Transformation is by using a 2×3 matrix.
```
